```python
del_cols=['job_id','location','department','salary_range','description','title','company_profile','benefits','requirements']
```

```python
df=df.drop(del_cols,axis=1)
```

```python
df
```

```python
col_list=list(df.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
#relacing null values
for i in col_list:
    if df[i].dtype=='object':
        df[i]=df[i].replace(np.nan,df[i].mode()[0],regex=True)
    
```

```python
df.isnull().sum()
```

```python
#lab encoding converting label into numerical format
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for i in col_list:
    if df[i].dtype=='object':
        df[i]=le.fit_transform(df[i])
```

```python
df
```

```python
y=df['fraudulent']
```

```python
#drop fradulent to use it as label
x=df.drop(['fraudulent'],axis=1)
```

```python
#divide the data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train) 
x_test = sc.transform(x_test)
```

```python
#naive bayes classifier
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

```python
y_pred  =  classifier.predict(x_test)
```

```python
naive_df=pd.DataFrame({'actual':y_test,
                    'predicted':y_pred})
```

```python
naive_df
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
ac = accuracy_score(y_test,y_pred)
```
